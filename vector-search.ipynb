```python
from os.path import sep

from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_openai import OpenAIEmbeddings
from langchain_chroma import Chroma
from torch.ao.quantization.backend_config.onednn import embedding_op_dtype_configs
```

```python
from dotenv import load_dotenv

load_dotenv()
```

```python
import pandas as pd

books = pd.read_csv("books_cleaned.csv")

books
```

```python
books["tagged_description"].to_csv("tagget_description.txt",
                                   lineterminator='\n',
                                   index = False,
                                   header = False)
```

```python
raw_documents = TextLoader("tagget_description.txt", encoding="utf-8").load()
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=0, separator="\n")
documents = text_splitter.split_documents(raw_documents)
```

```python
db_books =  Chroma.from_documents(
    documents,
    embedding=OpenAIEmbeddings())
```
